```python
df = df.dropna()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['gender'] = le.fit_transform(df['gender'])
df['address'] = le.fit_transform(df['address'])
df['work'] = le.fit_transform(df['work'])
df['Mjob'] = le.fit_transform(df['Mjob'])
df['Fjob'] = le.fit_transform(df['Fjob'])
df['activities'] = le.fit_transform(df['activities'])
df['friends'] = le.fit_transform(df['friends'])
df['health'] = le.fit_transform(df['health'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = df.drop(['habit'], axis=1)
y = df['habit']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
acc = accuracy_score(y_pred,y_test)
acc
```
